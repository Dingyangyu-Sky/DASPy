# DASPy Example

DAS data processing

## 1. Reading and visulization

```python
from daspy import read


sec = read() # input the path to the DAS file, here we use the example
print(sec)
sec.plot()
```

## 2. Frequency attribute

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(3, 1, figsize=(6,9))
sec.plot(ax=ax[0], obj='waveform', xmode='channel', tmode='origin', transpose=True, vmax=0.05)
sec.plot(ax=ax[1], obj='spectrogram', tmode='origin', kwargs_pro=dict(noverlap=156))
sec.plot(ax=ax[2], obj='fk')
plt.tight_layout()
```

## 3. Convert to velocity

```python
# integrate to strain
# bandpass filtering with detrending and taper
sec.time_integration().bandpass(1, 5, detrend=True, taper=0.05)


# add turning points into section (detected by DASPy)
sec.turning_channels = [2624, 2689, 2825, 2933]
sec.fk_rescaling(fmax=(5, 6))
sec.plot()
```

## 3. Low-quality channel checking

```python
import numpy as np
import scipy.io as scio
from daspy.advanced_tools.channel import channel_checking
from daspy.basic_tools.visualization import plot


data = scio.loadmat('../data/Ridgecrest_traffic_noise.mat')['Traffic_noise_figure_4'].T
good_chn, bad_chn, energy, threshold = channel_checking(data, deg=10, verbose=True)

fig, ax = plt.subplots(2, 1, figsize=(8,7), sharex=True)

# waveform
plot(data, fs=250, ax=ax[0], colorbar=False)

# energy
ax[1].plot(energy, lw=1)
ax[1].plot(threshold, c='k', ls='--', lw=1)

# dotted lines indicating the boundary of bad channels
idx = np.where(np.diff(bad_chn) > 1)[0]
idx = np.array([0] + idx.tolist() + (idx + 1).tolist() + [len(bad_chn) - 1])
ax[0].vlines(bad_chn[idx], 0, 15, color='r', lw=0.8, ls='--')
ax[1].vlines(bad_chn[idx], 0, threshold[bad_chn[idx]], color='r', lw=0.8, ls='--')

ax[1].set_ylim([0, 8.5])
ax[1].set_xlabel('Channel number')
ax[1].set_ylabel('Energy ($log_{10}$)')

plt.tight_layout()
plt.show()
```
